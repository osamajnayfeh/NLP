```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# stopwords.words('arabic')
# nltk.download('stopwords')
```

```python
text = "Hello, how are you doing today? I hope you're having a great day!"
sentences = nltk.sent_tokenize(text)
print(f"sentences: {sentences}")
tokens = nltk.word_tokenize(text)
print(f"tokens: {tokens}")
```

```python
arabic_text = "مرحبًا، كيف حالك اليوم؟ أتمنى أن يكون يومك رائعًا!"

word = word_tokenize(arabic_text)
print(f"arabic tokens: {word}")

Stop_Words = set(stopwords.words('arabic'))
print(f"arabic stop words: {Stop_Words}")

filtered_words = [w for w in word if not w in Stop_Words]
print(f"filtered arabic words: {filtered_words}")

filtered_text = ' '.join(filtered_words)
print(f"filtered arabic text: {filtered_text}")
```

```python
english_text = "This is an example of English text that we will use to demonstrate stop word removal."

words = word_tokenize(english_text)
print(f"english tokens: {words}")

Stop_Words = set(stopwords.words('english'))

filtered_words = [w for w in words if not w in Stop_Words]
print(f"filtered english words: {filtered_words}")

filtered_text = ' '.join(filtered_words)
print(f"filtered english text: {filtered_text}")
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
words = ["running", "jogging", "happily", "happiness"]
```

```python
for word in words:
    stemmed_word = stemmer.stem(word)
    print(f"{word} -> {stemmed_word}")
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
text = "The striped bats are hanging on their feet for best"
```

```python
# stemming
stemmer = PorterStemmer()
words = word_tokenize(text)
word_stemmed = [stemmer.stem(word) for word in word_tokenize(text)]
print(f"Stemmed words: {word_stemmed}")
```

```python
# lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]
print(f"Lemmatized words: {lemmatized_words}")
```
