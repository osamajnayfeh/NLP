```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# stopwords.words('arabic')
# nltk.download('stopwords')
```

```python
text = "Hello, how are you doing today? I hope you're having a great day!"
sentences = nltk.sent_tokenize(text)
print(f"sentences: {sentences}")
tokens = nltk.word_tokenize(text)
print(f"tokens: {tokens}")
```

```python
arabic_text = "مرحبًا، كيف حالك اليوم؟ أتمنى أن يكون يومك رائعًا!"

word = word_tokenize(arabic_text)
print(f"arabic tokens: {word}")

Stop_Words = set(stopwords.words('arabic'))
print(f"arabic stop words: {Stop_Words}")

filtered_words = [w for w in word if not w in Stop_Words]
print(f"filtered arabic words: {filtered_words}")

filtered_text = ' '.join(filtered_words)
print(f"filtered arabic text: {filtered_text}")
```

```python
english_text = "This is an example of English text that we will use to demonstrate stop word removal."

words = word_tokenize(english_text)
print(f"english tokens: {words}")

Stop_Words = set(stopwords.words('english'))

filtered_words = [w for w in words if not w in Stop_Words]
print(f"filtered english words: {filtered_words}")

filtered_text = ' '.join(filtered_words)
print(f"filtered english text: {filtered_text}")
```
